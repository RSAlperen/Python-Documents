#### Etiket Eklemek

```python
plt.title("Pie Chart")

slices = [60, 40]
labels = ["A", "B"]

plt.pie(slices, labels=labels)

plt.show()
```

#### Kenarlarına Ayraç (Seperator) Ekleme

```python
plt.title("Pie Chart")

slices = [60, 40]
labels = ["A", "B"]

plt.pie(slices, labels=labels, wedgeprops={"edgecolor":"black"})
plt.tight_layout()

plt.show()
```

#### Dilimlerin Renklerini Değiştirmek

```python
plt.title("Pie Chart")

slices = [60, 40, 15, 10]
labels = ["A", "B", "C", "D"]
colors = ["blue", "red", "yellow", "purple"]

plt.pie(slices, labels=labels, colors=colors, wedgeprops={"edgecolor":"black"})
plt.tight_layout()

plt.show()
```

#### Dilimleri Explode ile Vurgulamak

```python
most_used_lang = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
most_users = [59219, 55466, 47544, 36443, 35917]
```

```python
plt.title("Pie Chart")
explode = [0, 0, 0, 0.2, 0]

plt.pie(most_users, labels=most_used_lang, explode=explode, wedgeprops={"edgecolor":"black"})

plt.tight_layout()

plt.show()
```

#### Gölge Eklemek

```python
plt.title("Pie Chart")
explode = [0, 0, 0, 0.2, 0]

plt.pie(most_users, labels=most_used_lang, explode=explode, shadow=True, wedgeprops={"edgecolor":"black"})

plt.tight_layout()

plt.show()
```

#### Dilimleri Döndürme

```python
plt.title("Pie Chart")
explode = [0, 0, 0, 0.2, 0]

plt.pie(most_users, labels=most_used_lang, explode=explode, startangle=70, wedgeprops={"edgecolor":"black"})

plt.tight_layout()

plt.show()
```

#### Dilimlerin Yüzdesini Gösterme

```python
plt.title("Pie Chart")
explode = [0, 0, 0, 0.2, 0]

plt.pie(most_users, labels=most_used_lang, explode=explode, autopct="%1.1f%%", wedgeprops={"edgecolor":"black"})

plt.tight_layout()

plt.show()
```
