#### Splitting

```python
sensor_grup = df.groupby(["SENSOR_TYPE"])
```

```python
sensor_grup
```

```python
# Sadece "SENSOR_TYPE" sütunundan AUTOMATED_WEATHER_SENSOR olanları döndürür
sensor_grup.get_group("AUTOMATED_WEATHER_SENSOR")
```

**Yapılan şey bununla aynı:**

```python
filt = df["SENSOR_TYPE"] == "AUTOMATED_WEATHER_SENSOR"
```

```python
df.loc[filt]
```

#### Fonksiyon Uygulamak
**Bölgelere Göre Bölünmüş Median Sıcaklık** 

```python
bolge_group = df.groupby(["OBSERVATORY_NAME"])
```

```python
bolge_group["AVERAGE_TEMPERATURE"].median()
```

```python
bolge_group["AVERAGE_TEMPERATURE"].median().sort_values(ascending = False)
```

#### agg()
Birden çok fonksiyon uygular 

```python
bolge_group["AVERAGE_TEMPERATURE"].agg(["median", "mean"])
```
