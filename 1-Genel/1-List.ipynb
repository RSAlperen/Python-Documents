# List'in sonuna eleman ekleme
## append()
* İçine yazılan elemanı `list`'in sonuna ekler

```python
l = [1, 2, 3]
l.append(61)
l
```

## extend()
* İçine yazılan birden çok elemanı `list`'in sonuna ekler

```python
l = [1, 2, 3]
l.extend([100,200,500])
l
```

## insert()
* Spesifik bir `index`'e eleman eklemek için kullanılır. 

```python
l = [10, 20, 30]
l.insert(1, 50)
l
```

## remove()
* Belirli bir `elemanı` silmek için kullanılır. Eğer girilen eleman listede yok ise hata verir

```python
l = [10, 20, 30]
l.remove(20)
l
```

## pop()
* Listenin belirli bir `index`'indeki elemanı silip o elemanı **return** yapar

```python
l = [10, 20, 30]
l.pop(2)
```

## count()
* İçine yazdığımız değerin listede kaç defa gördüğünü döner

```python
l = [1, 2, 3, 3, 5]
l.count(3)
```

## Önemli notlar
* tmp = [1, 2, 3] #3 elemanlı bir liste oluştu. 
* eg = temp dediğimde `tmp` listenin elemanlarını kopyalamaz yada aktarmaz. Sadece `tmp` listesinin adresini eşitler. Böylece `tmp` listesinde görünen elemanlar `eg` listesinde de görünür. Fakat eşitlemeden sonra `tmp`listesinin elemanları güncellenirse aynı şekilde `eg`listesinin elemanlarıda güncellenir. Çünkü `eg`'de elemanlar değil `tmp`listesinin referansı var. Bir listeyi kopyalamak için **.copy()** fonksiyonu kullanılmalıdır. Böylece listenin adresi değil doğrudan elemanları kopyalanmış olur.

```python
tmp = [1, 2, 3]
eg = tmp

print(eg)

tmp[0:3] = [10, 20, 30]

print(eg)
```

```python
tmp = [1, 2, 3]
eg = tmp.copy()
print(eg)
tmp[0:3] = [10, 20, 30]
print(eg)
```
